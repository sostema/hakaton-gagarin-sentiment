# Технологичность

1. Изначально для получения sentence embneddings а нашей модели мы использовали подход основанный на ruT5Large, но позже мы нашли лидерборд по русскоязычным энкодерам и перешли на paraphrase-multilingual-mpnet-base-v2, что позволило ускорить модель в 25 раз.

https://github.com/avidale/encodechka/tree/master

2. В настоящее время LLM не являются лидерами в решении задач NER и Sentiment analysis (https://paperswithcode.com/task/named-entity-recognition-ner), но их можно использовать для быстрой разметки данных под эти задачи со специфическими отраслевыми инструкциями, объединив это преимущество с выданными авторами инструкциями по разметке датасета мы смогли подобрать промпт и доразметить 20000 дополнительных триплетов (corpus, entity, sentiment_score) за 20 минут используя фреймворк vvlm.

3. Изначально на выходе нашей модели стояли линейные слои, но после мы заменили их на решающие деревья, что дало нам не только прирост в качестве, но и в скорости модели на CPU более чем в 2 раза. Ну и позволило перейти на noGPU инференс :)

4. В качестве доразметки (не было применено из-за другого принципа разметки на лб) был написан скрипт для более глубокой доразметки данных, подав на вход текст сообщения и промежуток времени когда оно было написано мы вычленяем все сущности (компании) с помощью NER модели и обращаемся к апи мосбиржи (дав на вход тикет и временной промежуток) чтобы получить изменения цен торгов и оценить степень влияния данного сообщения.

5. Сделали упор на то, что в деревьях проще использовать айдишники компаний вместо суммирования эмбеддингов. Перестали считать эмбеддинги для названий компаний и взяли вместо них IssuerID, что позволило снизить вычислительную стоимость и ускорить модель.
